```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
```

```python
url = "https://www.basketball-reference.com/playoffs/NBA_2023.html"
```

```python
import pandas as pd

tables = pd.read_html(url)
```

```python
playoff23stats = tables[16]
playoff23stats.head()
```

```python
import seaborn as sns
```

```python
sns.barplot(x='Tm', y ='2P%', data=playoff23stats)
```

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')
```

```python
playoff23_2pts = playoff23stats[['Tm', '2P%']]
```

```python
playoff23_2pts
```

```python
playoff23_2pts.plot(kind='bar',x = "Tm", y = "2P%", figsize=(10, 6))

plt.xlabel('Team') # add to x-label to the plot
plt.ylabel('2pts Percentage') # add y-label to the plot
plt.title('2pts percentages in the playoffs 2023') # add title to the plot

plt.annotate('',  # s: str. will leave it blank for no text
             xy=(5, 0.56),  # place head of the arrow at point (year 2012 , pop 70)
             xytext=(12, 0.56),  # place base of the arrow at point (year 2008 , pop 20)
             xycoords='data',  # will use the coordinate system of the object being annotated
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
             )

# Annotate Text
plt.annotate('Nuggets killing the mid-range',  # text to display
             xy=(5, 0.57),  # start the text at at point (year 2008 , pop 30)
             rotation= 0,  # based on trial and error to match the arrow
             va='bottom',  # want the text to be vertically 'bottom' aligned
             ha='left',  # want the text to be horizontally 'left' algned.
             )

plt.show()
```
